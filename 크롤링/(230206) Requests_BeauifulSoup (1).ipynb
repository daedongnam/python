```python
def parsing_li(li):
  stock_name = li.find_all('a')[0].text
  title = li.find_all('a')[1].text
  price = li.find_all('div',attrs = {"class": 'line3'})[0].text
  price = price.replace('\n', '').replace('\t', '').replace('\r', '').replace(',', '').split(' ')[1][:-1]
  opinion = li.find_all('div',attrs = {"class": 'line3'})[1].text
  opinion = opinion.replace('\n', '').replace('\t', '').replace('\r', '')
  securities_company = li.find_all('div',attrs = {"class": 'line3'})[2].text
  date = li.find_all('div',attrs = {"class": 'line3'})[3].text

  return [stock_name, title, price, opinion, securities_company, date]
```
