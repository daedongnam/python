### 문제
- Q1. df_krx의 Marcap(시가총액)를 활용하여 시가총액 순위를 새로운 칼럼으로 정의 (칼럼명 : Marcap_rank)
- Q2. 순위데이터를 활용하여 100위 이하는 large-cap, 100~300위는 mid-cap, 이외 순위는 small-cap으로 새로운 칼럼 선언(칼럼명 : Marcap_size)
- Q3. Marcap_size(large-cap, mid-cap,small-cap) 수익률의 평균을 bar chart로 표현하세요

```python
df_krx = fdr.StockListing('KRX')
df_krx.head()
```

```python
#1. df_krx의 Marcap(시가총액)를 활용하여 시가총액 순위를 새로운 칼럼으로 정의 (칼럼명 : Marcap_rank)
df_krx['Marcap_rank'] = df_krx['Marcap'].rank(ascending = False)
df_krx.head()
```

```python
#2. 순위데이터를 활용하여 100위 이하는 large-cap, 100~300위는 mid-cap, 이외 순위는 small-cap으로 새로운 칼럼 선언(칼럼명 : Marcap_size)
def cap_size(data):
    if data < 100:
        return 'large-cap'
    elif data >= 100 and data < 300:
        return 'mid-cap'
    else:
        return 'small-cap'
    
df_krx['Marcap_size'] = df_krx['Marcap_rank'].apply(cap_size)
df_krx.head()
```

```python
#3.  Marcap_size(large-cap, mid-cap, small-cap) 수익률의 평균을 bar chart로 표현하세요
```

```python
y1 = df_krx[df_krx['Marcap_size']=='large-cap']['ChagesRatio'].mean()
y2 = df_krx[df_krx['Marcap_size']=='mid-cap']['ChagesRatio'].mean()
y3 = df_krx[df_krx['Marcap_size']=='small-cap']['ChagesRatio'].mean()
```

```python
#0. 데이터
labels = ['large-cap', 'mid-cap', 'small-cap']
xticks = np.arange(len(labels))
x = xticks
y = [y1, y2, y3]

#1. 생성
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xticks(xticks)
ax.set_xticklabels(labels)

#3.
ax.bar(x, y)
```
